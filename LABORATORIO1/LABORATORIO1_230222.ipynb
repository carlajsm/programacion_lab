### Ahora realizamos un método de series para crear un gráfico de la función con diferentes valores de "x"

```python
def func_exponente(x, n):
    
    e_to_x = 0
    for i in range(n):
        e_to_x = e_to_x + x**i/math.factorial(i)
        
    return e_to_x
```

```python
n_series = int(input('¿Cuántas series desea generar?: ')) 
n = int(input('Ingrese n de la serie : '))  

resultados_aproximados = []
resultados_reales = []

valores_x = np.linspace(1,20,n_series) #PARA NO DIGITAR X TANTAS VECES, LE DAREMOS UN RANGO 

for i in range(n_series):
    
        #x = int(input('Ingrese número x : '))
    #valores_x.append(x)
    print('valor de x es : ', valores_x[i])
    serie_resultado = func_exponente(valores_x[i], n)
    print('valor aprox ', serie_resultado)
    
    resultados_aproximados.append(serie_resultado)
    #print('mi lista de resultados aproximados es ', resultados_aproximados)
    
    valor_real = math.exp(valores_x[i])
    print('valor real ', valor_real)
    
    resultados_reales.append(valor_real)
    #print('mi lista de resultados reales es ', resultados_reales)
    
    print('==================')
```

```python
plt.plot(valores_x, resultados_aproximados,label='aproximada')
plt.legend()

plt.plot(valores_x, resultados_reales, '.',label = 'real')
plt.legend()
```

```python
def func_taylor(x, n):
    
    
    
    cos_aproximacion = 0
    
    for i in range(n):
        
        coef = (-1)**i
        print('coef : ', coef)
        print('i ' , i)
        num = x**(2*i)
        denom = math.factorial(2*i)
        print('el factorial denominador =  ', denom)
        cos_aproximacion = cos_aproximacion + (( coef ) * ( (num)/(denom) ))
        print("la aproximacion con i = ",i,"    La aproximación es : ", cos_aproximacion)
        print('=====================')
    
    return cos_aproximacion
```

```python
angulo = int(input('Ingrese ángulo : '))
x_angulo = (math.radians(angulo))

print("El ángulo",angulo,"tranformado en radianes es: ",x_angulo)

n = int(input('Ingrese el valor n: '))
resultado = func_taylor(x_angulo, n)
valor_real_coseno = math.cos(x_angulo)

print("Tomando como referencia el valor real de la funcion cos(x) = ",valor_real_coseno)
print(" ")
print("Para la serie con n = ", n,"      El resultado de la aproximacion de taylor es: ",resultado)
print(" ")
print("Observamos que mientras mas se incrementa la serie, la aproximación es mas cercana al valor real")
```
